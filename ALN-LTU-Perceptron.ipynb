### Biological Neurons

Artificial neural networks were orginally designed from biological neurons and to understand their learning and inference mechanism, might be of value to revisit how our human neurons work. 

Biological neurons main component is a __cell body__ where the nucleus and other important structures are contained. Departing from it, complex extensions, the __axons__, expand outside the body, reaching other neurons by comming in close contact with other cells. Just at the terminals of each axon, neural terminals, know as __telodendria__, point towards the direction of other cells bodies or terminals, via which they can communicate with each other. Finally, at the very end of this connections, the __synaptic terminals__ are met. Here, electro chemical compounds, know as __neurotransmitters__ are found in different proportions, defining how well a conection is made between different cells.

<img src="imgs/bnn.png" alt="Drawing" style="width: 400px;"/>

Communication between cells happen in form of __electric currents__. Whenever a neuron receive a stimulus from different organs, a difference in potential is generated between the wall-structure of the cell boy and the outside of the neuron, creating a small current that is denominated __action potential__. This current travels through the axon towards the synaptic terminals, where will be propapated depending on the amount of neurotransmitter that is found between each cells. It has been observed, that action potentials travelling through cell junctions are also the cause for the generation of more neurotransmitters within the connection. A phenomenon that forms part of what is known as __reinforcing the neural path__, and plays an important role in the __learning process__.


Even when the behaviour of individual neurons is regarded as simple, the overall addition and contribution of millions of these cells connected within a network are the constituents of the generation of real complex behaviours. ## The Linear Threshold Unit

After a series of advancements were realised in the world of computation and theory, In 1957 a new type of neuron was introduced. This new type of agent was able to receive numerical data as inputs and even to return a number that belonged to the real domain if it was told to do so. This type of neuron was given the name of __Linear Threshold Unit__, and was introduced by the american psychologist, Frank Rosenblatt, which is widely know for its work within the field of Artificial Intelligence. Later on, due to this properies, this new type of neuron opened a new world of posibilities and was the precursor of the simplest neural net - actually know as __the perceptron__. At would introduce a new meaning to theories previously utilised in the world of statistics. 


A __Logic Threshold Unit, (LTU)__ is a type of neuron with an architecture similar to what we observe nowadays. It has a series of inputs in form of real numbers, scaled via their associated weights. "Inside" the neuron, all these __scaled activations__ get aggregated numerically, creating a linear combination between the inputs and respective weights. After the aggregation has been calculated, a final activation function computes the final state of the neuron after the potentiation. This activation function comes in the form of thresholding functions, that is it will decide an activation or not depending on the input values. The __Step and sign functions__ are the most common types of activation functions for these agents, although any stepping function could be utilised. 

<img src="imgs/ltu.png" alt="Drawing" style="width: 350px;"/>

The LTU then could be visualised in the following manner:

Input weights __x1, x2, ..., Xn__ are scaled by some learned weight parameters __w0, w1, ..., Wn__, and aggregated through __Z__. Notice that the inputs go from 1 to N, while the weight parameters from 0 to 1. This because parameter __w0__ is equivalent to the bias term, __b__, of a linear combination; leading to:




$$ Z = w_0 + w_1 x_1 + w_2 x_2 + ... + w_n x_n $$

After aggregation gets completed, the final output of the perceptron is computed by the activation function $h_w(x)$:

In vector form:


$$ h_w(Z) = step(W^T X)$$

## The Power of LTUs

One of the great properties of __LTUs__ is their abilities to perform both __regression and classification__ tasks. Even most outstandingly, they are able to discover the type of task to be learn automatically from the data. 
Example, if our data is just a series of sequential continuous points, at the time of training - what is recognised as a __regression task__, our neural network is going to be able to fit its weights such that they will maximise the likelyhood of their prediction. 

In contrast, if the dataset implies a __classification task__ within its measured outputs, then the neural net will automatically fix the weights to do so. This is one of the most exciting results of their behaviour, nevertheless, nothing would be possible without its parameters training algorithms.

## Training through the Normal Equation 


All the similarities and the structure with a linear algebra problem, initially benefited the training process of the LTUs. If we observe carefully, a dataset for which a neuron is to be trained can be interpreted as a massive $mxn$ matrix $X$, which is to be multiplied with the input weights within a vector __W__. 

Also, if the output of the matrix for each observation is gathered within a new matrix __y__, then the perceptron inference algorithm just becomes the evaluation of the activation function of the linear equation $W^TX$ as seen before. So how to use that __linear algebra__ interpretation to fit out weights W?

If you recall, every training algorithm for almost every different type of supervised agent needs to be trained accompanied with a teacher - that is, a cost function. For neural networks, the first cost function to be utilised, was the __Mean Squared Error__, due to their great statistical and mathematical properties. However, more cost functions are now utilised depending on the __observed moments of data characteristics__ during the EDA to enhance their accuracy of prediction. 

Taking the __MSE__ as out teaching cost function depends on the training parameters in the following manner: $J(w)$. Then our error will just take the following shape:

$$ J(w) = \sum_{n=0}^{N-1}{(y_n - w^TX_n)}^2$$

If we want to minimise this cost function with respect to the parameters then it will just lead to the following result in vector form:

$$ J(w) = (y - Xw)^T(y - Xw)$$

$$ y y^T - y^TXw - (Xw)^Ty + (Xw)^TXw$$

Doing a bit of algebra:

$$ J(w) = y y^T - 2y^TXw + w^TX^TXw$$

Finding the maxima and minima:

$$ \frac{\delta J}{\delta w} = 0 $$ gets:

$$ - 2y^TX + 2X^TXw^T = 0$$

$$ X^TXw^T = y^TX$$

$$ (X^TX)^{-1}X^TXw^T = (X^TX)^{-1}y^TX$$

Leading to the final training closed-form result: 

$$ w = (X^TX)^{-1}y^TX$$

The previous equation is commonly known in the field of mathematics as __the Normal Equation__, and forms part of the first methodologies for the training of LUTs.

````python
class LinearThresholdUnitClassificator(BaseEstimator, TransformerMixin):
    
    def __init__(self, w = 1, b = 0):
        self.w = w #Class weight is given value of 1 or otherwise specified
        self.b = b #Class bias is given value of 0 or otherwise specified
        pass
    
    def fit(self, X1, X2, y1, y2, function = 'random'):
        
        '''
        X1: numpy array of shape: (#, )
        y1: numpy array of shape: (#, )
        X2: numpy array of shape: (#, )
        y2: numpy array of shape: (#, )
        function: string {'heaviside','tanh'}, if not specified will use random.randint()
        
        Obtain normal equation for classification 
        '''
        
        X1c = np.c_[X1, y1] #Concatenate X1 and y1 via columns
        X2c = np.c_[X2, y2] #Concatenate X2 and y2 via columns
        a = random.randint(1, 10) #Random int to use for new y1 classficcation
        b = random.randint(1, 10) #Random int to use for new y2 classficcation
        
        if(function == 'heaviside'):
            #Assign value of 1 to y1c & value of 0 to y2c
            y1c = np.zeros((len(X1c)))
            y2c = np.ones((len(X2c)))
        elif(function == 'tanh'):
            #Assign value of 1 to y1c & value of -1 to y2c
            y1c = np.ones((len(X1c)))
            y2c = -np.ones((len(X2c)))
        else:
            #Assign value of a to y1c & value of -b to y2c
            print(a), print(b)
            y1c = a*np.ones((len(X1c)))
            y2c = -b*np.ones((len(X2c)))
        
        Xt = np.r_[X1c, X2c] #Concatenate X1c and X2c via rows
        yt = np.r_[y1c, y2c] #Concatenate y1c and y2c via rows
        
        length = int(len(Xt)/2) #Get half of the array length
        
        fig = plt.figure()
        fig.add_subplot(111, projection='3d')
        
        if(function=='heaviside'):
            #Plot new Xcategorical & ycategorical
            plt.scatter(Xt[0:length, 0], Xt[0:length, 1], yt[0:length]+1)
            plt.scatter(Xt[length:, 0], Xt[length:, 1], yt[length:]+1)
        elif(function=='tanh'):
            #Plot new Xcategorical & ycategorical
            plt.scatter(Xt[0:length, 0], Xt[0:length, 1], yt[0:length])
            plt.scatter(Xt[length:, 0], Xt[length:, 1], yt[length:]+3)
        else:
            #Plot new Xcategorical & ycategorical
            plt.scatter(Xt[0:length, 0], Xt[0:length, 1], yt[0:length]+a)
            plt.scatter(Xt[length:, 0], Xt[length:, 1], yt[length:]+b+a)
            
        t1 = np.linalg.pinv(Xt.T.dot(Xt)) #Inverse matrix of dot product of X transverse and X (X'.X)^-1
        t2 = yt.T.dot(Xt) #Dot product of y transverse and X
        self.w = t1.dot(t2) 
        
        return self
    
    def predict(self, X1, X2, y1, y2):  
        '''
        Print line of best classification, through normal equation values.
        '''
        
        plt.plot(X1,y1,'*')
        plt.plot(X2,y2,'o')
        
        ss = np.linspace(-2, 4, 100)
        line_class = self.w[0]*ss + self.w[1]
        
        plt.plot(ss, line_class)
        print(self.w)
        
````
